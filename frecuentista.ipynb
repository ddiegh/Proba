```python
#bibliotecas que ocupamos 
import random
import matplotlib.pyplot as plt 
```

### Def. Probabilidad frecuentista

Sea $n_A$ el número de ocurrencias de un evento $A \in \mathcal{F}$ ($\sigma-albegra$) en $n$ realizaciones de un experimento aleatorio.

La probabilidad frecuentista del evento $A$ se define como:

$$P_{frec}(A) = \lim_{n \to \infty} \frac{n_A}{n}$$
Es decir, es la ocurrencia en las $n$ realizaciones, del evento que queremos observar

No podemos hacer el limite pero si probar el evento en una cantidad muy grande de ocaciones 

---

### Ejemplo

$\xi$: "lanzar una moneada equilibrada"

$\Omega = \{0, 1\}$ en donde $0$ representa obtener cruz y $1$ obtener cara.

---

Queremos calcular la probabilidad de obtener cara (que en teoría debería ser $1/2$).

Haciendo una simulación de este evento, obtenemos los siguientes resultados:
(observamos que los valores sí convergen a $1/2$)

```python
random.seed(42)
omega  = [0,1]  #los valores posibles 

n = 10000 #numero de iteraciones 
na = 0  #ocurrencias del evento que queremos 
valores = []  #las probas en cada iteracion

for i in range(1,n+1): #realizaremos el experimento n veces
    resultado = random.randint(0,1)    #escogemos un valor entre 0 y 1
    if resultado == 1:            #si es el valor que queremos, lo agregamos al contador de ocurrencias
        na += 1
    valores.append(na/i)    #hacemos en cociente entre la cantidad de ocurrencias y las iteraciones

#grafica
dom = range(1,n+1)
plt.figure(figsize=(12,10))
plt.plot(dom, valores, color='green', label = 'probabilidad frecuentista (obtener cara)')
plt.ylim(0,1)
plt.axhline(y=0.5, c='red', label = 'valor teorico (0.5)')
plt.legend()
plt.show()
```

### Ejemplo

$\xi$: "lanzar un dado equilibrado"

$\Omega = \{1, 2, 3, 4, 5, 6\}$ en donde cada valor representa la cara que se obtiene

---

Queremos calcular la probabilidad de obtener el $2$ (que en teoría debería ser $1/6$).

Haciendo una simulación de este evento, obtenemos los siguientes resultados:
(observamos que los valores sí convergen a $1/6$)

```python
random.seed(42)
omega  = [1,2,3,4,5,6]  #los valores posibles 

n = 10000 #numero de iteraciones 
na = 0  #ocurrencias del evento que queremos 
valores = []  #las probas en cada iteracion

for i in range(1,n+1): #realizaremos el experimento n veces
    resultado = random.randint(1,6)    #escogemos un valor entre 0 y 1
    if resultado == 1:            #si es el valor que queremos, lo agregamos al contador de ocurrencias
        na += 1
    valores.append(na/i)    #hacemos en cociente entre la cantidad de ocurrencias y las iteraciones

#grafica
dom = range(1,n+1)
plt.figure(figsize=(12,10))
plt.plot(dom, valores, color='green', label = 'probabilidad frecuentista (obtener un 2)')
plt.axhline(y=1/6, c='red', label = 'valor teorico (1/6)')
plt.legend()
plt.show()
```
